## Partie 2: Algorithme de Machine Learning

```python
import numpy as np
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./database.csv')
```

```python
data.head()
```

### Preprocessing

#### Supprimer les colonnes inutilisées

```python
droppedCols = ['Record Source', 'Victim Count', 'Perpetrator Count', 
               'Record ID', 'Agency Code', 'Agency Name', 'Agency Type',
              'City', 'State', 'Year', 'Month', 'Incident']
data = data.drop(droppedCols, axis = 1)

# Changer le type de la colonne Perpetrator Age
data['Perpetrator Age'] = pd.to_numeric(data['Perpetrator Age'], errors = 'coerce', downcast = 'integer')
```

```python
data.head()
```

#### Construire les cibles y

```python
# On va choisir le Crime Solved comme le target
y = data['Crime Solved'].values
unique_val, id_val = np.unique(y, return_inverse=True)
y[:] = id_val
y = y.astype(int)
y
```

#### Construire le X

```python
col_cat = [0, 2, 4, 5, 6, 8, 9, 10]
col_num = [3, 7]
```

##### Traiter les colonnes catégorielles

```python
X_cat = data.values[:, col_cat]
for col_id in range(len(col_cat)):
    unique_val, id_val = np.unique(X_cat[:, col_id], return_inverse=True)
    X_cat[:, col_id] = id_val
X_cat
```

```python
X_cat.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

X_cat = OneHotEncoder().fit_transform(X_cat).toarray()
X_cat.shape
```

##### Traiter les colonnes numériques

```python
X_num = data.values[:, col_num]
X_num = X_num.astype(float)

from sklearn.preprocessing import Imputer
imp_num = Imputer(missing_values = np.nan, strategy="mean")
X_num = imp_num.fit_transform(X_num)
X_num
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_num)
X_num = scaler.transform(X_num)
X_num
```

```python
X_num.shape
```

##### Construire X

```python
X = np.concatenate((X_num, X_cat), axis=1)
X.shape
```

```python
y.shape
```

### Appliquer les algorithmes

### Créer les classifieurs

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import time

clfs = {
    'NV': GaussianNB(),
    'CART': DecisionTreeClassifier(random_state = 1),
    'ID3': DecisionTreeClassifier(criterion='entropy', random_state = 1),
    'Stump': DecisionTreeClassifier(max_depth = 1),
    'MLP': MLPClassifier(hidden_layer_sizes=(5,)),
    'Bagging': BaggingClassifier(n_estimators=5),
    'AdaBoost': AdaBoostClassifier(n_estimators=5),
    'RF': RandomForestClassifier(n_estimators=5)
}
```

```python
def run_classifiers(clfs, X, y):
    kf = KFold(n_splits=2, shuffle=True, random_state=0)
    for i in clfs:
        clf = clfs[i]
        start = time.time()
        cv_acc = cross_val_score(clf, X, y, cv = kf)
        end = time.time()
        cv_auc = cross_val_score(clf, X, y, cv = kf, scoring="roc_auc")
        cv_pre = cross_val_score(clf, X, y, cv = kf, scoring="precision")
        print('Accuracy for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_acc), np.std(cv_acc)))
        print('Precision for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_pre), np.std(cv_pre)))
        print('ROC-AUC for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_auc), np.std(cv_auc)))
        print('Time for {0} is: {1:.3f}s'.format(i, end - start))
        print("=======================================")
```

```python
run_classifiers(clfs, X, y)
```

Conclusion, les algorithmes marchent assez bien sur cette base de données
