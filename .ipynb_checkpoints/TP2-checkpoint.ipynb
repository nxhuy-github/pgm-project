```python
import numpy as np
np.set_printoptions(threshold=np.nan,suppress=True)
import pandas as pd
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./credit_scoring.csv', sep=";")
print(df.head())
```

##### Transformer votre jeu de données et séparer ensuite les variables caractéristiques de la variable à prédire (status) en deux tableaux différents.

```python
data = df.values
X = data[:, :-1]
y = data[:, -1]
```

##### Séparer les données en deux partie: Training set et Test set

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)
```

##### Helper

```python
def execute(clf, X_train, y_train, X_test, y_test):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    
    acc_clf = accuracy_score(y_test, y_pred)
    pre_clf = precision_score(y_test, y_pred)
    rec_clf = recall_score(y_test, y_pred)
    
    return acc_clf, pre_clf, rec_clf
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
a, p, r = execute(knn, X_train, y_train, X_test, y_test)

accuracies.append(a)
precisions.append(p)
recalls.append(r)
```

```python
index = [('Decision Tree', '-'), ('KNN', '-')]
_t = pd.Series(accuracies, index=index,)
index = pd.MultiIndex.from_tuples(index)
_t = _t.reindex(index)
_tdf = pd.DataFrame({'Accuracy':_t, 'Precision': precisions, 'Recall': recalls}) 
print(_tdf)
```

### Pipeline

##### Helper

```python
from sklearn.base import BaseEstimator, TransformerMixin
class Norm_ACP(BaseEstimator, TransformerMixin):
    def __init__(self, n_components=3):
        self.n_components = n_components

    def fit(self, _X, _y):
        from sklearn.preprocessing import StandardScaler
        from sklearn.decomposition import PCA

        self.scaler = StandardScaler()
        self.scaler.fit(_X)
        
        self.pca = PCA(self.n_components)
        self.pca.fit(_X)
        return self

    def transform(self, _X):
        import numpy as np
        
        scaled = self.scaler.transform(_X)
        pca = self.pca.transform(_X)
        scaled_pca = np.concatenate((scaled, pca), axis = 1)

        return scaled_pca
```

##### Pipeline avec KNN

```python
from sklearn.pipeline import Pipeline

k = 3
knn = Pipeline((
    ('norm_acp', Norm_ACP(k)),
    ('knn', KNeighborsClassifier(n_neighbors = 5))
))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=1)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

_acc = accuracy_score(y_test, y_pred_knn)
print('Accuracy de KNN: ',_acc)
```

##### Pipeline avec Decision Tree

```python
tree_clf = Pipeline((
    ('norm_acp', Norm_ACP()),
    ('tree_clf', DecisionTreeClassifier(random_state=1))
))

tree_clf.fit(X_train, y_train)
y_pred_tree = tree_clf.predict(X_test)

_acc = accuracy_score(y_test, y_pred_tree)
print('Accuracy de Decision Tree: ', _acc)
```

### Comparaison de plusieurs algorithmes d’apprentissage

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import time

clfs = {
    'NV': GaussianNB(),
    'CART': DecisionTreeClassifier(random_state = 1),
    'ID3': DecisionTreeClassifier(criterion='entropy', random_state = 1),
    'Stump': DecisionTreeClassifier(max_depth = 1),
    'MLP': MLPClassifier(hidden_layer_sizes=(20,10)),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Bagging': BaggingClassifier(n_estimators=50),
    'AdaBoost': AdaBoostClassifier(n_estimators=50),
    'RF': RandomForestClassifier(n_estimators=50)
}

def run_classifiers(clfs, X, y):
    kf = KFold(n_splits=10, shuffle=True, random_state=0)
    for i in clfs:
        clf = clfs[i]
        start = time.time()
        cv_acc = cross_val_score(clf, X, y, cv = kf)
        end = time.time()
        cv_auc = cross_val_score(clf, X, y, cv = kf, scoring="roc_auc")
        cv_pre = cross_val_score(clf, X, y, cv = kf, scoring="precision")
        print('Accuracy for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_acc), np.std(cv_acc)))
        print('Precision for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_pre), np.std(cv_pre)))
        print('ROC-AUC for {0} is: {1:.3f} +/- {2:.3f}'.format(i, np.mean(cv_auc), np.std(cv_auc)))
        print('Time for {0} is: {1:.3f}s'.format(i, end - start))
        print("=======================================")
```

```python
run_classifiers(clfs, X, y)
```

```python
data = yeld.values
y_stars = data[:, 0]
y_stars = y_stars.astype('int')
X_text = data[:, 1]


X_text = pipeline_vectorizer.fit_transform(X_text.astype('U'))

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_text, y_stars, test_size=0.5, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 50)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score

acc_clf = accuracy_score(y_test, y_pred)
pre_clf = precision_score(y_test, y_pred, average='micro')
rec_clf = recall_score(y_test, y_pred, average='micro')
print(acc_clf, pre_clf, rec_clf)
```
